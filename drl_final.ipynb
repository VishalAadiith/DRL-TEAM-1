```python
import numpy as np
import pandas as pd
import yfinance as yf
from gym import Env, spaces
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from datetime import datetime, timedelta
import time
import talib

class EnhancedMultiStockPortfolioEnv(Env):
    def __init__(self, tickers, initial_balance=100000, lookback_window=30, risk_free_rate=0.02, target_volatility=0.2):
        super(EnhancedMultiStockPortfolioEnv, self).__init__()
        self.tickers = tickers
        self.initial_balance = initial_balance
        self.lookback_window = lookback_window
        self.risk_free_rate = risk_free_rate
        self.target_volatility = target_volatility  # Target volatility for risk adjustment

        self.action_space = spaces.Box(low=0, high=1, shape=(len(tickers),))
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(len(tickers) * 11 + 3,))

        self.reset()

    def reset(self):
        self.balance = self.initial_balance
        self.shares = {ticker: 0 for ticker in self.tickers}
        self.data = self._get_data()
        self.dates = self.data.index.tolist()
        self.current_step = self.lookback_window
        return self._next_observation()

    def _get_data(self):
        end_date = datetime.now().strftime('%Y-%m-%d')
        start_date = (datetime.now() - timedelta(days=365)).strftime('%Y-%m-%d')  # 1 year of data
        data = yf.download(self.tickers, start=start_date, end=end_date)

        # Calculate additional indicators
        for ticker in self.tickers:
            data[f'{ticker}_RSI'] = talib.RSI(data['Close'][ticker])
            data[f'{ticker}_SMA'] = talib.SMA(data['Close'][ticker], timeperiod=20)
            data[f'{ticker}_MACD'], _, _ = talib.MACD(data['Close'][ticker])

        return data.ffill().bfill()

    def _next_observation(self):
        obs = []
        current_date = self.dates[self.current_step]
        lookback_date = self.dates[self.current_step - self.lookback_window]

        for ticker in self.tickers:
            stock_data = self.data.loc[lookback_date:current_date]
            close_pct_change = stock_data['Close'][ticker].pct_change(fill_method=None).dropna()
            volume_pct_change = stock_data['Volume'][ticker].pct_change(fill_method=None).dropna()

            obs.extend([
                close_pct_change.mean(),
                close_pct_change.std(),
                volume_pct_change.mean(),
                stock_data['High'][ticker].max() / stock_data['Close'][ticker].iloc[-1] - 1,
                stock_data['Low'][ticker].min() / stock_data['Close'][ticker].iloc[-1] - 1,
                self.shares[ticker] * self.data['Close'][ticker].loc[current_date] / max(self.balance, 1),
                self.shares[ticker],
                stock_data[f'{ticker}_RSI'].iloc[-1] / 100,  # Normalized RSI
                stock_data[f'{ticker}_SMA'].iloc[-1] / stock_data['Close'][ticker].iloc[-1] - 1,  # SMA relative to current price
                stock_data[f'{ticker}_MACD'].iloc[-1],
                (stock_data['Close'][ticker].iloc[-1] - stock_data['Close'][ticker].iloc[0]) / stock_data['Close'][ticker].iloc[0]  # Price trend
            ])

        # Dynamic risk adjustment using portfolio volatility
        obs.extend([
            self.balance / self.initial_balance,
            self._calculate_portfolio_volatility(),
            self._calculate_risk_adjustment()  # New observation for dynamic risk adjustment
        ])
        return np.nan_to_num(obs)

    def _calculate_portfolio_volatility(self):
        current_date = self.dates[self.current_step]
        portfolio_value = sum(self.shares[ticker] * self.data['Close'][ticker].loc[current_date] for ticker in self.tickers) + self.balance
        if portfolio_value == 0:
            return 0
        weights = np.array([(self.shares[ticker] * self.data['Close'][ticker].loc[current_date]) / portfolio_value for ticker in self.tickers])
        returns = self.data['Close'].pct_change(fill_method=None).dropna().iloc[-252:]
        portfolio_return = returns.dot(weights)
        return portfolio_return.std() * np.sqrt(252)

    def _calculate_risk_adjustment(self):
        """Adjusts risk based on portfolio volatility."""
        volatility = self._calculate_portfolio_volatility()
        risk_adjustment_factor = min(1, self.target_volatility / max(volatility, 1e-8))
        return risk_adjustment_factor

    def _calculate_sharpe_ratio(self):
        current_date = self.dates[self.current_step]
        portfolio_value = sum(self.shares[ticker] * self.data['Close'][ticker].loc[current_date] for ticker in self.tickers) + self.balance
        if portfolio_value == 0:
            return 0
        weights = np.array([(self.shares[ticker] * self.data['Close'][ticker].loc[current_date]) / portfolio_value for ticker in self.tickers])
        returns = self.data['Close'].pct_change(fill_method=None).dropna().iloc[-252:]
        portfolio_return = returns.dot(weights)
        portfolio_return_mean = portfolio_return.mean() * 252
        portfolio_volatility = portfolio_return.std() * np.sqrt(252)
        return (portfolio_return_mean - self.risk_free_rate) / max(portfolio_volatility, 1e-8)

    def _calculate_diversification(self):
        current_date = self.dates[self.current_step]
        portfolio_value = sum(self.shares[ticker] * self.data['Close'][ticker].loc[current_date] for ticker in self.tickers) + self.balance
        if portfolio_value == 0:
            return 0
        weights = np.array([(self.shares[ticker] * self.data['Close'][ticker].loc[current_date]) / portfolio_value for ticker in self.tickers])
        return 1 - np.sum(weights**2)  # Inverse of Herfindahl-Hirschman Index

    def step(self, actions):
        self.current_step += 1
        if self.current_step >= len(self.dates):
            return self._next_observation(), 0, True, {}

        current_date = self.dates[self.current_step]
        current_prices = self.data['Close'].loc[current_date]
        prev_portfolio_value = sum(self.shares[ticker] * current_prices[ticker] for ticker in self.tickers) + self.balance

        # Adjust actions by risk factor (dynamic risk adaptation)
        risk_adjustment_factor = self._calculate_risk_adjustment()
        actions = np.nan_to_num(actions) * risk_adjustment_factor

        actions_sum = np.sum(actions)
        if actions_sum > 0:
            actions = actions / actions_sum
        else:
            actions = np.ones_like(actions) / len(actions)

        for i, ticker in enumerate(self.tickers):
            desired_value = actions[i] * prev_portfolio_value
            current_value = self.shares[ticker] * current_prices[ticker]
            value_diff = desired_value - current_value

            if value_diff > 0:  # Buy
                shares_to_buy = value_diff // current_prices[ticker]
                cost = shares_to_buy * current_prices[ticker]
                if cost <= self.balance:
                    self.shares[ticker] += shares_to_buy
                    self.balance -= cost
            elif value_diff < 0:  # Sell
                shares_to_sell = min(abs(value_diff) // current_prices[ticker], self.shares[ticker])
                revenue = shares_to_sell * current_prices[ticker]
                self.shares[ticker] -= shares_to_sell
                self.balance += revenue

        new_portfolio_value = sum(self.shares[ticker] * current_prices[ticker] for ticker in self.tickers) + self.balance
        portfolio_return = (new_portfolio_value - prev_portfolio_value) / prev_portfolio_value
        sharpe_ratio = self._calculate_sharpe_ratio()
        diversification = self._calculate_diversification()
        cash_penalty = max(0, 0.2 - self.balance / new_portfolio_value)  # Penalty for holding more than 20% cash

        # Reward adjusted for dynamic risk
        reward = sharpe_ratio + diversification - cash_penalty + risk_adjustment_factor

        done = self.current_step >= len(self.dates) - 1
        return self._next_observation(), reward, done, {}


def train_model(env, total_timesteps=10000):  # Reduced timesteps for faster training
    model = PPO("MlpPolicy", env, verbose=1, tensorboard_log="./ppo_portfolio_tensorboard/")
    model.learn(total_timesteps=total_timesteps)
    return model

import matplotlib.pyplot as plt

def plot_individual_stocks(env_data, tickers, final_shares, buy_sell_signals):
    for ticker in tickers:
        stock_data = env_data['Close'][ticker]
        dates = env_data.index

        plt.figure(figsize=(12, 6))
        plt.plot(dates, stock_data, label=f'Stock Price: {ticker}', color='blue')

        # Plot buy signals
        buy_dates = buy_sell_signals[ticker]['buy']
        plt.scatter(buy_dates, stock_data[buy_dates], color='green', label='Buy', marker='^', s=100)

        # Plot sell signals
        sell_dates = buy_sell_signals[ticker]['sell']
        plt.scatter(sell_dates, stock_data[sell_dates], color='red', label='Sell', marker='v', s=100)

        plt.title(f'Stock Trading - Buy and Sell Points for {ticker}')
        plt.xlabel('Date')
        plt.ylabel('Stock Price')
        plt.legend()

        # Add text annotation for final shares
        final_price = stock_data.iloc[-1]
        final_share_count = final_shares[ticker]
        plt.annotate(f'Final Shares: {final_share_count}',
                     xy=(dates[-1], final_price),
                     xytext=(10, 10),
                     textcoords='offset points',
                     ha='left',
                     va='bottom',
                     bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
                     arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))

        plt.show()

def plot_portfolio_vs_sp500(initial_value, portfolio_values, dates, sp500_data):
    plt.figure(figsize=(10, 6))

    # Portfolio performance plot
    portfolio_returns = [((value / initial_value) - 1) * 100 for value in portfolio_values]
    plt.plot(dates, portfolio_returns, label="Portfolio Performance", color='blue')

    # S&P 500 performance plot
    sp500_initial = sp500_data[0]
    sp500_returns = [((price / sp500_initial) - 1) * 100 for price in sp500_data]
    plt.plot(dates, sp500_returns, label="S&P 500 Performance", color='red')

    plt.title('Portfolio vs. S&P 500 Performance')
    plt.xlabel('Date')
    plt.ylabel('Cumulative Return (%)')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    tickers = ['AAPL', 'GOOGL', 'MSFT', 'AMZN', 'META', 'TSLA', 'NVDA', 'JPM', 'JNJ', 'V']
    initial_balance = 100000

    start_time = time.time()
    env = DummyVecEnv([lambda: EnhancedMultiStockPortfolioEnv(tickers, initial_balance)])
    print(f"Environment creation time: {time.time() - start_time:.2f} seconds")

    start_time = time.time()
    model = train_model(env, total_timesteps=10000)  # Reduced timesteps
    print(f"Model training time: {time.time() - start_time:.2f} seconds")

    # Test the model for 1 month (approximately 21 trading days)
    obs = env.reset()
    done = False
    total_steps = 21

    initial_value = initial_balance
    portfolio_values = [initial_value]
    buy_sell_signals = {ticker: {'buy': [], 'sell': []} for ticker in tickers}
    previous_shares = env.get_attr('shares')[0]

    for _ in range(total_steps):
        action, _states = model.predict(obs)
        obs, rewards, done, info = env.step(action)

        # Calculate and store portfolio value at each step
        current_balance = env.get_attr('balance')[0]
        current_shares = env.get_attr('shares')[0]
        current_prices = env.get_attr('data')[0]['Close'].iloc[-1]
        current_portfolio_value = current_balance + sum(current_shares[ticker] * current_prices[ticker] for ticker in tickers)
        portfolio_values.append(current_portfolio_value)

        # Track buy and sell signals
        current_date = env.get_attr('data')[0].index[-1]
        for ticker in tickers:
            if current_shares[ticker] > previous_shares[ticker]:
                buy_sell_signals[ticker]['buy'].append(current_date)
            elif current_shares[ticker] < previous_shares[ticker]:
                buy_sell_signals[ticker]['sell'].append(current_date)

        previous_shares = current_shares.copy()

        if done:
            break

    env_data = env.get_attr('data')[0]
    final_balance = env.get_attr('balance')[0]
    final_shares = env.get_attr('shares')[0]
    final_prices = env_data['Close'].iloc[-1]

    final_portfolio_value = final_balance + sum(final_shares[ticker] * final_prices[ticker] for ticker in tickers)

    # Fetch S&P 500 data
    sp500 = yf.Ticker('^GSPC')
    sp500_data = sp500.history(start=env_data.index[0], end=env_data.index[-1])['Close']

    print(f"Initial Portfolio Value: ${initial_value:.2f}")
    print(f"Final Portfolio Value: ${final_portfolio_value:.2f}")
    print(f"Total Return: {((final_portfolio_value - initial_value) / initial_value) * 100:.2f}%")
    print(f"S&P 500 Return: {((sp500_data.iloc[-1] / sp500_data.iloc[0]) - 1) * 100:.2f}%")
    print("\nFinal Portfolio Composition:")
    for ticker in tickers:
        stock_value = final_shares[ticker] * final_prices[ticker]
        percentage = (stock_value / final_portfolio_value) * 100
        print(f"{ticker}: {final_shares[ticker]} shares (${stock_value:.2f}, {percentage:.2f}% of portfolio)")
    cash_percentage = (final_balance / final_portfolio_value) * 100
    print(f"Cash: ${final_balance:.2f} ({cash_percentage:.2f}% of portfolio)")

    # Plot individual stock prices with buy/sell signals and portfolio performance vs S&P 500
    plot_individual_stocks(env_data, tickers, final_shares, buy_sell_signals)
    plot_portfolio_vs_sp500(initial_value, portfolio_values, env_data.index[-len(portfolio_values):], sp500_data[-len(portfolio_values):])
```
